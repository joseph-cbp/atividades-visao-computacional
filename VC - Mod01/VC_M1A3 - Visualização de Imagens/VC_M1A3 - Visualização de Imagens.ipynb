### Lendo outra imagem

```python
image_rio = plt.imread("./rio.jpeg")
print(type(image_rio))
image_rio[0,0]
plt.imshow(image_rio)
```

### Mudando espaço de cores

```python
plt.imshow(image_rio[:, :, ::-1], cmap = 'Reds')
```

```python
plt.imshow(image_rio[:, :, 0], cmap='inferno')
```

```python
teste_image = image_rio.copy()
teste_image[:, :, 1] = 250
plt.imshow(teste_image)
```

### Imagens Geradas com NumPy

```python
ruido = np.random.rand(100, 100)
plt.imshow(ruido, cmap = "gray")
```

```python
ruido_azul = np.zeros((100, 100, 3))
ruido_azul[:, :, 2] = ruido * 5
plt.imshow(ruido_azul)
```
